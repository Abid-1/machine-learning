```python
import pandas as pd
```

```python
                                           #GETTING DTATA      
                                           # split into input and target values  (x,y)                     
```

```python
df = pd.read_csv("1000_Companies.csv")
```

```python
df.head()
```

```python
x=df.iloc[:,[0,1,2,3]].values
x
```

```python
y=df.iloc[:,[4]].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
x[:,3]=le.fit_transform(x[:,3])
x
```

# importing a library Sk learn

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Applying Ml Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

# Checking the Accuracy in Model

```python
accuracy =regressor.score(x_test,y_test)
print("Accuracy of the model is: ", accuracy)
```

# Creating an Array wih Prediction

```python
new_data = [10000,15000,50000,"Florida"]  # Example new data point
new_data[3] = le.transform([new_data[3]])[0]  # Transform the city name
predicted_value = regressor.predict([new_data]) 
predicted_value
```
